```python
# Import the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load Dataset

fifa_df = pd.read_csv("E:/EPL_20_21.csv")
```

```python
fifa_df.head()
```

```python
fifa_df.info()
```

```python
fifa_df.describe()
```

```python
fifa_df.isna().sum()
```

```python
# Create 2 new columns

fifa_df['MinsPerMatch'] = (fifa_df['Mins']/ fifa_df['Matches']).astype(int)
fifa_df['GoalsPerMatch'] = (fifa_df['Goals']/ fifa_df['Matches']).astype(float)
fifa_df.head()
```

```python
# Total Goals

Total_Goals = fifa_df['Goals'].sum()
print(Total_Goals)
```

```python
# Penalty Goals

Total_PenaltyGoals = fifa_df['Penalty_Goals'].sum()
print(Total_PenaltyGoals)
```

```python
# Penalty Attempts

Total_PenaltyAttempts = fifa_df['Penalty_Attempted'].sum()
print(Total_PenaltyAttempts)
```

```python
# Pie chart for penalties missed vs scored

plt.figure(figsize=(13,6))
pl_not_scored = fifa_df['Penalty_Attempted'].sum() - Total_PenaltyGoals
data=[pl_not_scored,Total_PenaltyGoals]
labels = ['Penalties missed','Penalties Scored']
color = sns.color_palette('husl',8)
plt.pie(data, labels = labels , colors = color, autopct='%.0f%%')
plt.show()
```

```python
# Unique Positions

fifa_df['Position'].unique()
```

```python
# Total FW players

fifa_df[fifa_df['Position'] == 'FW']
```

```python
# Players from Different Nations

np.size((fifa_df['Nationality'].unique()))
```

```python
# Most Players from Which countries

nationality = fifa_df.groupby('Nationality').size().sort_values(ascending = False)
nationality.head(10).plot(kind = 'bar', figsize=(12,6), color = sns.color_palette('viridis'))
```

```python
# Clubs with maximum players in their squad

fifa_df['Club'].value_counts().nlargest(5).plot(kind='bar', color=sns.color_palette("magma"))
```

```python
# Clubs with least players in their squad
fifa_df['Club'].value_counts().nsmallest(5).plot(kind = 'bar', color=sns.color_palette('tab10'))
```

```python
# Players based on age group
Under20 = fifa_df[fifa_df['Age']<=20]
age20_25 = fifa_df[(fifa_df['Age']>20) & (fifa_df['Age']<=25)]
age25_30 = fifa_df[(fifa_df['Age']>25) & (fifa_df['Age']<=30)]
Above30 = fifa_df[fifa_df['Age']>30]
```

```python
x=np.array([Under20['Name'].count(),age20_25['Name'].count(),age25_30['Name'].count(),Above30['Name'].count()])
mylabels = ["<=20",">20 & <=25",">25 &<=30", ">30"]
plt.title('Total Players with Age Groups',fontsize=20)
plt.pie(x, labels = mylabels , autopct="%.1f%%")
plt.show()
```

```python
# Total under 20 players in each club

players_under_20 = fifa_df[fifa_df['Age']<20]
players_under_20['Club'].value_counts().plot(kind = 'bar', color = sns.color_palette('magma') )
```

```python
# Under 20 players in ManU

players_under_20[players_under_20['Club']=='Manchester United']
```

```python
# Under 20 players in Chelsea

players_under_20[players_under_20['Club']=='Chelsea']
```

```python
# Average age of players in each club 

plt.figure(figsize=(12,6))
sns.boxplot(x='Club', y='Age', data= fifa_df)
plt.xticks(rotation=90)
```

```python
num_player = fifa_df.groupby('Club').size()
data = (fifa_df.groupby('Club')['Age'].sum()/ num_player)
data.sort_values(ascending = False)
```

```python
# Total Assist from each club

Assits_by_clubs = pd.DataFrame(fifa_df.groupby('Club',as_index=False)['Assists'].sum())
sns.set_theme(style="whitegrid",color_codes=True)
ax = sns.barplot(x='Club',y='Assists',data=Assits_by_clubs.sort_values(by="Assists"),palette='cubehelix')
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Assists",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of Clubs vs Total Assists',fontsize=20)
```

```python
# Top 10 Assists

top_10_assists = fifa_df[['Name','Club','Assists','Matches']].nlargest(n=10, columns='Assists')
top_10_assists
```

```python
Goals_by_clubs = pd.DataFrame(fifa_df.groupby('Club',as_index=False)['Goals'].sum())
sns.set_theme(style="whitegrid",color_codes=True)
ax = sns.barplot(x='Club',y='Goals',data=Goals_by_clubs.sort_values(by="Goals"),palette='Set2')
ax.set_xlabel("Club",fontsize=30)
ax.set_ylabel("Goals",fontsize=20)
plt.xticks(rotation=75)
plt.rcParams["figure.figsize"]=(20,8)
plt.title('Plot of Clubs vs Total Goals',fontsize=20)
```

```python
# Most goals by players

top_10_goals = fifa_df[['Name','Club','Goals','Matches']].nlargest(n=10, columns='Goals')
top_10_goals
```

```python
# Goals per match

top_10_goals_per_match = fifa_df[['Name','GoalsPerMatch','Matches','Goals']].nlargest(n=10, columns='GoalsPerMatch')
top_10_goals_per_match
```

```python
# Pie chart - Goals with assist and without assist

plt.figure(figsize=(14,7))
assists = fifa_df['Assists'].sum()
data=[Total_Goals - assists, assists]
labels = ['Goals without assists', 'Goals with assists']
color = sns.color_palette('Set2')
plt.pie(data, labels=labels, colors = color, autopct='%.0f%%')
plt.show()
```

```python
# Top 10 players with most yellow cards

fifa_yellow=fifa_df.sort_values(by='Yellow_Cards', ascending=False)[:10]
plt.figure(figsize=(20,6))
plt.title("Players with the most yellow cards")
c=sns.barplot(x=fifa_yellow['Name'],y=fifa_yellow['Yellow_Cards'],label ='Players', color='yellow')
plt.ylabel('Number of Yellow Cards')
c.set_xticklabels(c.get_xticklabels(),rotation=45)
c
```
